```python
import os
from PIL import Image
import cv2
def read_images():
    orginal_images = []
    modified_images = []
    path = "C:\\Users\\furkan\\Desktop\\images"
    for filename in os.listdir(path):
        #file name i√ßinde orginal varsa orginal iamge ekle yoksa modified image ekle
        if "original" in filename:
            img = Image.open(os.path.join(path, filename))
            orginal_images.append(cv2.imread(os.path.join(path, filename)))
        else:
            img = Image.open(os.path.join(path, filename))
            modified_images.append(cv2.imread(os.path.join(path, filename)))
    return orginal_images , modified_images

orginal_images , modified_images = read_images()
```

```python
def draw_histograms(orginal_images, modified_images):
    import matplotlib.pyplot as plt

    for i in range(len(orginal_images)):
        image_list = ["A", "B", "C", "D", "E", "F"]
        orginal_hist = cv2.calcHist([orginal_images[i]], [0], None, [256], [0, 256])
        modified_hist = cv2.calcHist([modified_images[i]], [0], None, [256], [0, 256])
        
        plt.figure()
        plt.title(f"{image_list[i]}_Histograms")
        plt.xlabel("Bins")
        plt.ylabel("# of Pixels")
        plt.plot(orginal_hist)
        plt.plot(modified_hist)
        plt.legend()
        plt.show()

draw_histograms(orginal_images, modified_images)
```

```python
import matplotlib.pyplot as plt
A_orginal = orginal_images[0]
A_modified = modified_images[0]

A_modified_new = cv2.bitwise_not(A_orginal)

plt.figure()
plt.title("A Orginal Image")
plt.imshow(cv2.cvtColor(A_orginal, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure()
plt.title("A Modified Image")
plt.imshow(cv2.cvtColor(A_modified, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure()
plt.title("A Re-Modified Image")
plt.imshow(cv2.cvtColor(A_modified_new, cv2.COLOR_BGR2RGB))
plt.show()
```

Histogram Processing

```python
B_orginal = orginal_images[1]
B_modified = modified_images[1]

B_orginal = cv2.cvtColor(B_orginal, cv2.COLOR_BGR2GRAY)
B_modified_new = cv2.equalizeHist(B_orginal)

plt.figure()
plt.title("B Orginal Image")
plt.imshow(cv2.cvtColor(B_orginal, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure()
plt.title("B Modified Image")
plt.imshow(cv2.cvtColor(B_modified, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure()
plt.title("B Re-Modified Image")
plt.imshow(cv2.cvtColor(B_modified_new, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
C_original = orginal_images[2]
C_modified = modified_images[2]
import numpy as np

#avg blur
kernel = np.ones((5, 5), np.float32)/25
C_modified_new = cv2.filter2D(C_original, -1, kernel)
for i in range(5):
    C_modified_new = cv2.filter2D(C_modified_new, -1, kernel)

plt.figure()
plt.title("C Original Image")
plt.imshow(C_original)
plt.show()

plt.figure()
plt.title("C Modified Image")
plt.imshow(C_modified)
plt.show()

plt.figure()
plt.title("C Re-Modified Image")
plt.imshow(cv2.cvtColor(C_modified_new, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
D_original = orginal_images[3]
D_modified = modified_images[3]

kernel = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])
D_modified_new = cv2.filter2D(D_original, -1, kernel)

plt.figure()
plt.title("D Original Image")
plt.imshow(D_original)

plt.figure()
plt.title("D Re-Modified Image")
plt.imshow(D_modified_new)

plt.figure()
plt.title("D Modified Image")
plt.imshow(cv2.cvtColor(D_modified_new, cv2.COLOR_BGR2RGB))
```

```python
E_original = orginal_images[4]
E_modified = modified_images[4]

E_modified_new = cv2.copyMakeBorder(E_original, 2, 2, 2, 2, cv2.BORDER_CONSTANT, value=0)

E_modified_new = cv2.medianBlur(E_modified_new, 5)
for i in range(10):
    E_modified_new = cv2.medianBlur(E_modified_new,5)

plt.figure()
plt.title("E Original Image")
plt.imshow(E_original)

plt.figure()
plt.title("E Modified Image")
plt.imshow(E_modified)

plt.figure()
plt.title("E Re-Modified Image")
plt.imshow(cv2.cvtColor(E_modified_new, cv2.COLOR_BGR2RGB))
```

```python
F_original = orginal_images[5]
F_modified = modified_images[5]

alpha = 1.7
beta = 20
F_modified_new = cv2.convertScaleAbs(F_original, alpha=alpha, beta=beta)

plt.figure()
plt.title("F Original Image")
plt.imshow(F_original)

plt.figure()
plt.title("F Modified Image")
plt.imshow(F_modified)


plt.figure()
plt.title("F Re-Modified Image")
plt.imshow(cv2.cvtColor(F_modified_new, cv2.COLOR_BGR2RGB))
```
